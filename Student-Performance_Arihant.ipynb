## Data Imputation and Data Encoding

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# separating columns for impute-encode pipeline
numerical_features = [i for i in df if df[i].dtype != 'object']
categorical_features = [i for i in df if df[i].dtype == 'object']

# numerical values are imputed by fitting null with mean
numeric_transformer = SimpleImputer(strategy='mean')

# categorical values are imputed by fitting null with mode, and then ordinally encoded to a numerical value
categorical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),('label_encoder', OrdinalEncoder())])

# column-wise preprocessing unit working on each column parallely
preprocessor = ColumnTransformer(transformers=[('num', numeric_transformer, numerical_features),('cat', categorical_transformer, categorical_features)])

# pipeline to sequentialize above units to single flow
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

processed = pipeline.fit_transform(df)

df_processed = pd.DataFrame(processed, columns=numerical_features + categorical_features)
```

## Outlier detection

```python
from scipy.stats import zscore

z_score = {} # holds entire framework of z-score tested outliers

for i in df_processed.columns: # iterating over columns
    z_scores = zscore(df_processed[i]) # z score of current column
    outliers = z_scores[abs(z_scores) > 2.5] # judging threshold of z-score is 2.5

    z_score[i] = outliers # storing outliers based on columns
```

```python
IQR_score = {} # holds entire framework of IQR tested outliers

for column in df_processed.columns:
    Q1 = df_processed[column].quantile(0.25) # 1st Quartile
    Q3 = df_processed[column].quantile(0.75) # 3rd Quartile
    
    IQR = Q3 - Q1 # difference of quartiles : equivalent to boxplot's central distribution mass

    outliers = df_processed[(df_processed[column] < Q1-1.5*IQR) | (df_processed[column] > Q3+1.5*IQR)][column] # judging values greater than 3rduartile and below 1st quartile as outliers 

    if not outliers.empty:
        IQR_score[column] = outliers # storing outliers
```

## Problem type: Regression

since we have observed Exam_Score is a continuous value feature, we treat the problem as a regression problem to predict based on parameters what the Exam_Score for a new child will be based on current data.

```python
df.columns
```

```python
x = df_processed.copy().drop(columns=['Exam_Score'], axis = 1)
y = df_processed.copy()['Exam_Score']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.15)

print("x:", x.shape)
print("y:", y.shape)
print("x_train:", x_train.shape)
print("x_test:", x_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

from sklearn.metrics import mean_absolute_error, root_mean_squared_error

predictions = model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Root Mean Square Error:", rmse)
```

```python
from keras.models import Sequential
from keras.layers import InputLayer, Conv1D, MaxPooling1D, Flatten, Dense
from keras.callbacks import ModelCheckpoint

model = Sequential()
model.add(InputLayer(input_shape=(19, 1)))
model.add(Conv1D(filters=16, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='nadam', loss='mse', metrics=['mae'])

model_checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss',verbose=1, save_best_only=True, mode='min')

model.summary()
```

```python
history = model.fit(x_train, y_train, validation_split=0.15, epochs=20, verbose=1, batch_size=40, callbacks= model_checkpoint)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Plot Loss vs Val Loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()

# Plot MAE vs Val MAE
plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Mean Absolute Error vs Validation MAE')
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# final fitted model

from sklearn.metrics import mean_absolute_error, root_mean_squared_error

predictions = model.predict(x_test)

mae = mean_absolute_error(y_test, predictions)
rmse = root_mean_squared_error(y_test, predictions)

print("Mean Absolute Error:", mae)
print("Root Mean Square Error:", rmse)
```

## Problem type: Classification

since we have observed Exam_Score is a continuous value feature, we can convert it into a Classification problem by binning Exam_Score into, say, Grade or Result:

Grade:

* Grade A: Exam_Score ≥ 85

* Grade B: 70 ≤ Exam_Score < 85

* Grade C: 55 ≤ Exam_Score < 70

* Grade D: 40 ≤ Exam_Score < 55

* Grade F: Exam_Score < 40

Result:

Pass: Marks > 45

Fail: Marks <= 45

```python
def categorize_exam_score(score):
    if score >= 85:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 55:
        return 'C'
    elif score >= 40:
        return 'D'
    else:
        return 'F'

df_processed['Grade'] = df_processed['Exam_Score'].apply(categorize_exam_score)
```

```python
set(df_processed['Grade'])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df_processed['Grade'] = encoder.fit_transform(df_processed['Grade'])
```

```python
set(df_processed['Grade'])
```

```python
x = df_processed.copy().drop(columns=['Grade'], axis = 1)
y = df_processed.copy()['Grade']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.15)

print("x:", x.shape)
print("y:", y.shape)
print("x_train:", x_train.shape)
print("x_test:", x_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import InputLayer, Conv1D, MaxPooling1D, Flatten, Dense
from keras.callbacks import ModelCheckpoint

model = Sequential()
model.add(InputLayer(input_shape=(20, 1)))
model.add(Conv1D(filters=16, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='nadam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model_checkpoint = ModelCheckpoint('cat_best_model.keras', monitor='val_accuracy',verbose=1, save_best_only=True, mode='max')

model.summary()
```

```python
history = model.fit(x_train, y_train, validation_split=0.15, epochs=20, verbose=1, batch_size=40, callbacks= model_checkpoint)
```

```python
# final fitted model

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

y_pred = model.predict(x_test)

y_pred_classes = np.argmax(y_pred, axis=1)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred_classes)
precision = precision_score(y_test, y_pred_classes, average='weighted')
recall = recall_score(y_test, y_pred_classes, average='weighted')
f1 = f1_score(y_test, y_pred_classes, average='weighted')

# Print metrics
print("\nMetrics:")
print("Accuracy: {:.2f}%".format(accuracy * 100))
print("Precision (weighted): {:.2f}%".format(precision * 100))
print("Recall (weighted): {:.2f}%".format(recall * 100))
print("F1 Score (weighted): {:.2f}%".format(f1 * 100))

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_classes))

# Confusion matrix
conf_mat = confusion_matrix(y_test, y_pred_classes)

print(conf_mat)
```
